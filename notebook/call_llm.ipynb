# Call LLM using Google LLM API

```python
from langchain_google_genai import ChatGoogleGenerativeAI

# Initialize the Gemini model
llm = ChatGoogleGenerativeAI(model="gemma-4-31b-it")
```

```python
response = llm.invoke("Why LangChain is popular?")
print(response.text)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

# Define the conversation messages
messages = [
    SystemMessage("You are a helpful assistant that answers in short one line."),
    HumanMessage("Why LangChain is popular?")
]

# Invoke the model
response = llm.invoke(messages)
print(response.content) 
```

# Call LLM using Groq Models

```python
import os
# os.getenv('GROQ_API_KEY')
```

```python
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq(model="qwen/qwen3-32b", reasoning_format="parsed")
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

# Define the conversation messages
messages = [
    SystemMessage("You are a helpful assistant that answers in short one line."),
    HumanMessage("Why LangChain is popular?")
]

# Invoke the model
response = llm.invoke(messages)
print(response.content) 
```
