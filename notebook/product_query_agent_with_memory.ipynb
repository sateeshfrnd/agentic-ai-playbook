```python
from langchain.tools import tool
@tool
def get_product_details(name: str) -> str:
    """Look up a product by name and return its price, rating, and description."""
    p = PRODUCTS.get(name.lower())
    if not p:
        return f"Product not found. Available: {', '.join(PRODUCTS)}"
    return str(p)
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)
```

```python
from langchain.agents import create_agent
agent = create_agent(
    llm,
    tools=[get_product_details],
    system_prompt="You are a helpful product assistant for an online tech store.",
)
```

```python
def ask(question: str):
    result = agent.invoke({"messages": [{"role": "user", "content": question}]})
    print(result["messages"][-1].content)
```

```python
@tool
def get_product_reviews(product_name: str) -> str:
    """Get reviews for a product name."""
    reviews = REVIEWS.get(product_name.lower())
    if not reviews:
        return f"No reviews found for {product_name}. Available products: {', '.join(REVIEWS)}"
    return str(reviews)


@tool
def get_average_review_rating(product_name: str):
    """Get average review rating for a product name."""
    reviews = REVIEWS.get(product_name.lower())
    if not reviews:
        return f"No reviews found for {product_name}. Available products: {', '.join(REVIEWS)}"
    
    avg = sum(r["rating"] for r in reviews) / len(reviews)
    return round(avg, 2)
```

```python
from langgraph.checkpoint.memory import MemorySaver

memory = MemorySaver()

agent2 = create_agent(
    llm,
    tools=[get_product_details, get_product_reviews, get_average_review_rating],
    system_prompt=(
        "You are a helpful assistant for our tech store. ONLY answer questions using the tools. "
        "Do NOT invent products. If the user asks about a generic product or category (like smartwatch or headphones), "
        "always use the tool to query our store's specific matching products (e.g., 'smart watch' or 'wireless headphones') first."
    ),
    checkpointer=memory,
)
```

```python
# Create a config with thread_id for conversation memory persistence
config = {"configurable": {"thread_id": "product_store_session"}}

def ask2(question: str):
    result = agent2.invoke({"messages": [{"role": "user", "content": question}]}, config=config)
    print(result["messages"][-1].content)
```

```python
ask2("What do people say about smart watchs?")
```
